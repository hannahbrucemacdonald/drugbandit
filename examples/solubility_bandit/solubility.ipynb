Bayesian bandit for efficient calculation of the most soluble ligand of a set of eight benzene derivatives.

Solvation free energy has a normal distribution of unknown expectation and variance, with a normal-inverse gamma congugate prior.

Reward of each bandit is calculated
- sampling from the experimental gaussian distribution
- sampling using short Yank simulation

```python
from drugbandit import bandit, pull, choose
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import random

%matplotlib inline
```

```python
def plotbanditsvsexperiment(bandits):
    '''
    Plots the Bandit's expected free energy and error against the experimental free energy and error
    '''
    experimental = [arm.fe for arm in bandits]
    experimental_error = [arm.error for arm in bandits]

    calculated = [arm.nu for arm in bandits]
    calculated_error = [arm.sigma for arm in bandits]

    plt.errorbar(experimental,calculated,xerr=experimental_error,yerr=calculated_error,fmt='o',ms=0,color='k',alpha=0.3)
    plt.scatter(experimental,calculated,color=colors)
    plt.plot(experimental,experimental,color='b',alpha=0.2)
    plt.xlabel('Experimental / kcal mol$^{-1}$')
    plt.ylabel('Calculated / kcal mol$^{-1}$')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    return
```

```python
# used for printing output
BOLD = '\033[1m'
END = '\033[0m'
```

**Loading the ligands**

```python
bandits = []
colors = ['b','g','r','c','m','y','purple','orange']

with open('set_of_eight.smiles','r') as f:
    index = 0
    for line in f:
        if line[0] != '#':
            details = line.split(';')
            name, exp_fe, exp_error = details[2], float(details[3]), float(details[4])
            bandits.append(bandit.Bandit(name,exp_fe,exp_error,index,color=colors[index]))
            index += 1
```

**Sampling from experimental distribution**

*e-greedy algorithm*

```python
bandit_steps = 1000
for step in range(1,bandit_steps+1):
    # e-greedy, with e = 0.1
    pick_bandit = choose.e_greedy(bandits, 1.,reverse=True)
    reward = pull.Pull(pick_bandit)
    pick_bandit.update(reward)
    
    if step % 250 == 0 or step == 1:
        print('{}Step {} {}'.format(BOLD,step,END))
        ax = plt.subplot(1, 1, 1)
        for arm in bandits:
            arm.plot()
        handles, labels = ax.get_legend_handles_labels()
        lgd = ax.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
        plt.show()
        
        plotbanditsvsexperiment(bandits)
        
        for arm in bandits:
            arm.print_bandit(step)
```

**Sampling from experimental distribution**

*boltzmann algorithm*

```python
bandits = []

colors = ['b','g','r','c','m','y','purple','orange']

with open('set_of_eight.smiles','r') as f:
    index = 0
    for line in f:
        if line[0] != '#':
            details = line.split(';')
            name, exp_fe, exp_error = details[2], float(details[3]), float(details[4])
            bandits.append(bandit.Bandit(name,exp_fe,exp_error,index,color=colors[index]))
            index += 1


bandit_steps = 1000
for step in range(1,bandit_steps+1):
    # Boltzmann sampling 
    pick_bandit = choose.boltzmann(bandits,5,reverse=True)
    reward = pull.Pull(pick_bandit)
    pick_bandit.update(reward)
    
    if step % 250 == 0 or step == 1:
        print('{}Step {} {}'.format(BOLD,step,END))
        ax = plt.subplot(1, 1, 1)
        for bandit in bandits:
            bandit.plot()
        handles, labels = ax.get_legend_handles_labels()
        lgd = ax.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
        plt.show()
        
        plotbanditsvsexperiment(bandits)
        
        for bandit in bandits:
            bandit.print_bandit(step)
```

**Plotting the proportion of pulls for each ligand**

```python
x = [bandit.fe for bandit in bandits]
y = [(bandit.nsteps*100)/bandit_steps for bandit in bandits]

plt.scatter(x,y,color=colors)
plt.xlabel('hydration free energy / kcal mol$^{-1}$')
plt.ylabel('proportion of simulation time / %')
plt.show()
```
